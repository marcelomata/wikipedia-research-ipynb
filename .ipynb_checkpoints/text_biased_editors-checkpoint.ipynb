```python
import csv
import json
import pandas as pd
from pprint import pprint
import os
import sys
csv.field_size_limit(sys.maxsize)
```

```python
# Open lists of biased editors with row indices names
edfile = '/home/michael/school/cprose_research/wp/wp_articles/citation_bias_editors.csv'
cite = pd.read_csv(edfile, index_col=1)
cite
```

```python
# Read in talk page data
talk = pd.read_csv('/home/michael/school/cprose_research/wp/wikipedia/data/talk/ipc_utf8_talkpages.csv', index_col=0)
talk
```

```python
# Split into quantiles, get top Israeli and Palestinian
quantiles = pd.qcut(cite['normalized_citations'], 10, range(10))
isr = quantiles[quantiles==9].index.tolist()
pal = quantiles[quantiles==0].index.tolist()
print(len(isr))
print(len(pal))
```

```python
# Split into pos/neg I/P
isr = cite[cite['normalized_citations'] > 0].index.tolist()
pal = cite[cite['normalized_citations'] < 0].index.tolist()
print(len(isr))
print(len(pal))
```

```python
# Get I/P biased talk page text
isrtalk = talk[talk['username'].isin(isr)]
print(len(isrtalk))
paltalk = talk[talk['username'].isin(pal)]
print(len(paltalk))
```

```python
biasedtalk = pd.concat([isrtalk, paltalk], keys=['isr', 'pal'])
#label = 'posneg'
label = '10'

# Write biased csv
biasedtalk.to_csv('/home/michael/school/cprose_research/wp/wp_articles/biased_talk_{:s}.csv'.format(label), index_label=['bias', 'talk_page_revision_id'])
```

```python
# Open JSON files of editor citation bias lists
filepath = '/home/michael/school/cprose_research/wp/wikipedia/data/editor_citationbias_novandal.json'
with open(filepath) as f:
    citelist = json.load(f)
print("No vandal, normalized")
pprint(citelist[:5])
pprint(citelist[-5:])
print()
```

```python
# Build DataFrame
cite = pd.DataFrame([[c[0], c[1][0], c[1][1]] for c in citelist if c[1][1]!=0], columns=['name', 'sum_citations', 'normalized_citations'])
print(len(cite))
print(len(citelist))
cite
```

```python
# Get top quantiles from citation bias list
nquantiles = 20
cite['citation_bias_category'] = pd.qcut(cite['normalized_citations'], nquantiles, labels=range(nquantiles))
cite
```

```python
# Write citation bias csv
cite.loc[:, ['name','normalized_citations', 'sum_citations']].to_csv('/home/michael/school/cprose_research/wp/wp_articles/citation_bias_editors.csv')
```

```python
# Build Pal list
pal = cite[cite['citation_bias_category']==0]['name'].tolist()
print(pal[0])
len(pal)
```

```python
# Build Israeli list
isr = cite[cite['citation_bias_category']==19]['name'].tolist()
print(isr[0])
len(isr)
```

```python
# Collect text from Palestinian, Israeli top editors
isr_rows = []
pal_rows = []

rawdirpath = '/home/michael/school/cprose_research/wp/wp_articles/extract/extract_articles/ipc_articles_raw'
for fname in os.listdir(rawdirpath):
    fpath = os.path.join(rawdirpath, fname)
    
    with open(fpath) as f:
        r = csv.reader(f, delimiter='\t')
        for row in r:
            text = row[2]
            ed = row[3]
            if ed in isr:
                isr_rows.append(row)
            elif ed in pal:
                pal_rows.append(row)
    
    print("Finished with {:s}".format(fpath))
                
print(len(isr_rows))
print(len(pal_rows))
```
