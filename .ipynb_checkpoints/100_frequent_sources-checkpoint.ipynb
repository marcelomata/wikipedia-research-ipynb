```python
import csv
import os
import re, sys
from collections import defaultdict, Counter
from operator import itemgetter
import pandas as pd
import math
import urllib.request
from IPython.core.debugger import Tracer; debug_here = Tracer()
```

```python
# Initialize for aggregating citations for top sources (also a script) or editor counts (should be made a script)
#citations = Counter()
#adds = Counter()
#dels = Counter()
ec = defaultdict(Counter) # ec[name][citation]
srcs = set()
citedir = '/home/michael/school/research/wp/wp_articles/citediffs/'
#outpath = '/home/michael/school/research/wp/wp_articles/citations_add_del.csv'
ecmatpath = '/home/michael/school/research/wp/wp_articles/editor_citations_matrix.csv'
fps = []
for p in sorted(os.listdir(citedir)):
    if not p.startswith('.'): fps.append(os.path.join(citedir, p))

# Aggregate citations from citediffs
for fp in fps:
    with open(fp) as f:
        rows = [row for row in csv.DictReader(f)]
        
    for i, row in enumerate(rows):
        if row['add/delete'] == 'add':
            if row['vandalism_comment'] == '0':
                #adds[row['citation']] += 1
                #citations[row['citation']] += 1
                ec[row['editor']][row['citation']] += 1
                srcs.add(row['citation'])
        elif row['add/delete'] == 'delete':
            if row['vandalism_comment'] == '0':
                #dels[row['citation']] += 1
                #citations[row['citation']] += 1
                ec[row['editor']][row['citation']] -= 1
    
srcs = sorted(list(srcs))

#print(len(adds))
#print(len(dels))
#print(len(citations))
print(len(ec))
print(len(srcs))
```

```python
# Get 100 most frequent citations
freqpath = '/home/michael/school/research/wp/wikipedia/data/citations_annotated.csv'
freq = pd.read_csv(freqpath)
#crit = freq['name'].map(lambda x: isinstance(x, str))
#freq100 = freq[crit]
freq100 = freq.loc[:99,:]

freqsrcs = {}
for _,row in freq100.iterrows():
    if isinstance(row['variants'], str):
        freqsrcs[row['name']] = [row['citation']]+ [var[1:-1] for var in row['variants'][1:-1].split(', ')]
    else:
        freqsrcs[row['name']] = [row['citation']]
len(freqsrcs)
```

```python
# Get bot list
bots = []
cnum = ""
done = False

# Get bot category page
category_base_url = "https://en.wikipedia.org/w/api.php?" + \
    "action=query&" +  \
    "list=categorymembers&" + \
    "format=json&" + \
    "cmtitle=Category%3A{:s}&" +  \
    "cmprop=title&" + \
    "cmlimit=500" + \
    "&cmcontinue={:s}"
    
while not done:
    category_url = category_base_url.format(urllib.parse.quote("All_Wikipedia_bots"), cnum)
    #print(category_url)

    page =  urllib.request.urlopen(category_url).read().decode('utf-8')
    #print(page)

    # Extract bot names
    botsp = re.findall(r'"User\:(?P<username>[\w\ /\.]*)"\}', page)
    botsp = [re.sub(r'/.*', '', ed) for ed in botsp]
    #print(botsp[:10])
    bots += botsp
#     print(len(bots))

    # Get continue number
    #cnum = re.search(r'cmcontinue":"(?P<cnum>[^"]*)"', page)
    cnum = re.search(r'cmcontinue":"([^"]*)"', page)
    if cnum:
        cnum = cnum.group(1)
#         print(cnum)
    else:
        done = True

# Add other bots
other_bots = ['SmackBot']
bots += other_bots
len(bots)
```

```python
# Print out editor citation matrix
ec100 = {}
variants = [x for sublist in [val for val in freqsrcs.values()] for x in sublist]

# Reconstruct a combined one
for ed in [e for e in ec if e not in bots and e != '']:
    for src in ec[ed]:
        if src in variants:
            # Find which is name
            selected_variants = [val for val in freqsrcs.values() if src in val][0]
            name = list(freqsrcs.keys())[list(freqsrcs.values()).index(selected_variants)]
            #if not ed in ec100: ec100[ed] = Counter()
            if not ed in ec100: ec100[ed] = {}
            if not name in ec100[ed]:
                ec100[ed][name] = ec[ed][src]
            else:
                ec100[ed][name] += ec[ed][src]

ecmat100 = pd.DataFrame(ec100)

# Write editor-citation matrix
ecmat100.to_csv(ecmatpath, na_rep=0)

ecmat100
```

```python
# Print input for SVDFeature--sources and editors sorted alphabetically
ratings = []
for i, ed in enumerate(sorted(ec100.keys())):
    for j, src in enumerate(sorted(freqsrcs.keys())):
        if src in ec100[ed]:
            rating = [ec100[ed][src], 0, 1, 1, "{:d}:{:d}".format(i+1,1), "{:d}:{:d}".format(j+1,1)]
            ratings.append(rating)
    
print(len(ratings))

csvpath = '/home/michael/school/research/wp/wp_articles/editor_citation_svdfeature.txt'

with open(csvpath, 'w') as f:
    w = csv.writer(f, delimiter=' ')
    for r in ratings:
        w.writerow(r)
```

```python
# Aggregate added, deleted citations for each revision
citations = Counter()
adds = Counter()
dels = Counter()
citedir = '/home/michael/school/research/wp/wp_articles/citations/'
cite_diffdir = '/home/michael/school/research/wp/wp_articles/citediffs/'
infopath = '/home/michael/school/research/wp/wikipedia/data/citations_reverts.csv'
fps = []
for p in sorted(os.listdir(citedir)):
    if not p.startswith('.'): fps.append(os.path.join(citedir, p))

# Load article revision metadata
with open(infopath) as f:
    r = csv.DictReader(f)
    info = {(row['article_name'], row['timestamp']): row for row in r}

# Parse added, deleted citations for each revision
#for fp in fps[1:2]:
for fp in fps:
    print('Parsing citations in {:s}'.format(fp), end='')
    sys.stdout.flush()
    with open(fp) as f:
        rows = [row for row in csv.DictReader(f)]
        
    #for i, row in enumerate(rows[beg:end]):
    prev_ts = rows[0]['timestamp']
    prev_cites = Counter()
    cites = Counter()
    outrows = [['article_name', 'timestamp', 'editor', 'add/delete', 'citation', 'edit_comment', 
               'edit_length', 'revert_md5', 'revert_comment', 'vandalism_comment']]
    for i, row in enumerate(rows):
        if i % 50000 == 0:
            #print(i/1000, 'k')
            print('.', end='')
            sys.stdout.flush()
            
        ts = row['timestamp']
        ri = info[(row['article_name'], ts)]
        
        # Update
        if ts != prev_ts: 
            d = prev_cites-cites
            a = cites-prev_cites
            for c in a:
                outrows.append([row['article_name'], ts, ri['editor'], 'add', c, ri['edit_comment'], ri['edit_length'],
                               ri['revert_md5'], ri['revert_comment'], ri['vandalism_comment']])
            for c in d:
                outrows.append([row['article_name'], ts, ri['editor'], 'delete', c, ri['edit_comment'], ri['edit_length'],
                               ri['revert_md5'], ri['revert_comment'], ri['vandalism_comment']])
            
            prev_cites = cites
            cites = Counter()
            prev_ts = ts

        # Extract cite book
        if re.search(r'(?:c|C)ite book', row['citation']):
            m_first = re.search(r'(?:c|C)ite book.*first=(?P<first>.*?)\ *?\|', row['citation'])
            m_last = re.search(r'(?:c|C)ite book.*last=(?P<last>.*?)\ *?\|', row['citation'])
            m_author = re.search(r'(?:c|C)ite book.*author=(?P<author>.*?)\ *?\|', row['citation'])
            m_title = re.search(r'(?:c|C)ite book.*title=(?P<title>.*?)\ *?\|', row['citation'])
            if m_first and m_last:
                cites[' '.join([m_first.group('first'), m_last.group('last')])] += 1
            elif m_last:
                cites[m_last.group('last')] += 1
            elif m_author:
                cites[m_author.group('author')] += 1
            elif m_title:
                cites[m_title.group('title')] += 1

        # Extract cite web
        elif re.search(r'(?:c|C)ite ?web', row['citation']):
            m_auth = re.search(r'(?:c|C)ite ?web.*?(?:author=(?P<author>.*?))(?:\||$)', row['citation'])
            m_url = re.search(r'(?:c|C)ite ?web.*?url=http:\/\/(?:www\.)?(?P<url>.*?)\/', row['citation'])
            # r'(?:c|C)ite web.*(?:author=(?P<author>.*?)\|.*)?.*url=http:\/\/(?:www\.)?(?P<url>.*?)\/'
            if m_auth: 
                #cites[('web', m_auth.group('author'))] += 1
                cites[m_auth.group('author')] += 1
            if m_url: 
                cites[m_url.group('url')] += 1
                #cites[('web', m_url.group('url'))] += 1

        # Extract cite journal
        elif re.search(r'(?:c|C)ite ?journal', row['citation']):
            m_auth = re.search(r'(?:c|C)ite ?journal.*?(?:author=(?P<author>.*?))(?:\||$)', row['citation'])
            m_journal = re.search(r'(?:c|C)ite ?journal.*?(?:journal=(?P<journal>.*?))(?:\||$)', row['citation'])
            #if m_auth: cites[('journal', m_auth.group('author'))] += 1
            if m_auth: cites[m_auth.group('author')] += 1
            #if m_journal: cites[('journal', m_journal.group('journal'))] += 1
            if m_journal: cites[(m_journal.group('journal'))] += 1

        # Extract cite news
        elif re.search(r'(?:c|C)ite ?news', row['citation']):
            m_auth = re.search(r'(?:c|C)ite ?news\*?(?:author=(?P<author>.*?)\|.*)(?:\||$)', row['citation'])
            m_url = re.search(r'(?:c|C)ite ?news.*?url=http:\/\/(?:www\.)?(?P<url>.*?)\/', row['citation'])
            #if m_auth: cites[('news', m_auth.group('author'))] += 1
            #if m_url: cites[('news', m_url.group('url'))] += 1
            #if m_url: citations.append(('web', m_url.group('url')))
            if m_auth: cites[m_auth.group('author')] += 1
            if m_url: cites[m_url.group('url')] += 1

        # Extract just url
        elif re.search(r'http:\/\/(?:www\.)?(?P<url>.*?)[ \/]', row['citation']):
            c = re.search(r'http:\/\/(?:www\.)?(?P<url>.*?)[ \/]', row['citation']).group('url')
            #citations.append(("url", c))
            cites[c] += 1
            
        # Extract harvnb
        elif re.search(r'[Hh]arvnb\|(.*?)\|', row['citation']):
            c = re.search(r'[Hh]arvnb\|(.*?)\|', row['citation']).group(1)
            cites[c] += 1

        # Extract harvtxt
        elif re.search(r'[Hh]arvtxt\|(.*?)\|', row['citation']):
            c = re.search(r'[Hh]arvtxt\|(.*?)\|', row['citation']).group(1)
            cites[c] += 1

        else: # Default
            m = re.search(r'\ *(?:"|\'\'|\[\[)?(.+?)(?:\.|,|"|“|\'|\||\])', row['citation'])
            if m:
                c = m.group(1)
                if 'http' in c:
                    debug_here()
                #print(beg+i+2, 'default', c)
                #citations.append(('unknown', c))
                cites[c] += 1
            #else:
            #    print("no parse")
            
    # Write citation diffs
    with open(os.path.join(cite_diffdir, os.path.split(fp)[1][:-4]) + '_citediff.csv', 'w') as f:
        w = csv.writer(f)
        w.writerows(outrows)
            
    print("finished")
```
