```python
# Build our old IPC article list
old_ipc_articles = []
csvpath = '/home/michael/school/cprose_research/wp/ipc_articles_2016-01-21.tsv'

with open(csvpath, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        article = row[0]
        if not article in old_ipc_articles:
            old_ipc_articles.append(article)
            
len(old_ipc_articles)
```

```python
# Check to see if any Wp category articles aren't in our set
category_not_ours = []
for article in ipc_wp_category:
    if article not in old_ipc_articles:
        category_not_ours.append(article)
        
print(len(category_not_ours))
```

```python
# Write the exceptions out to a file
wp_ipc_list_csvpath = '/home/michael/school/cprose_research/wp/new_articles_2016-01-22.csv'
with open(wp_ipc_list_csvpath, 'w') as csvfile:
    writer = csv.writer(csvfile)
    for article in category_not_ours:
        csvfile.write(article + '\n')
```
