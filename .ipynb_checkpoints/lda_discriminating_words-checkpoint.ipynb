```python
# Parse output from MALLET

n_topics = 15
bl_top_wds = [None] * n_topics # for every topic, dict (wd: freq)
for i in range(n_topics):
    bl_top_wds[i] = defaultdict(int)
    
malletpath = '/home/michael/software/mallet-2.0.8RC3/wp15_bl_toks.txt'

with open(malletpath, 'r') as malletfile:
    lines = malletfile.read().splitlines()[1:]
    for i, line in enumerate(lines):
        if i % 3 == 0:
            parts = line.split('\t')
            filename = parts[1]
        elif i % 3 == 1:
            parts = line.split(' ')[1:]
            for part in parts:
                item = part.split('/')
                topic = int(item[0])
                wd = item[1]
                bl_top_wds[topic][wd] += 1
        
len(bl_top_wds)
```

```python
# Split corpus to find top words in topics per perspective

n_topics = 15
top_wds_i = [None] * n_topics # for every topic, dict (wd: freq)
top_wds_p = [None] * n_topics # for every topic, dict (wd: freq)
for i in range(n_topics):
    top_wds_i[i] = defaultdict(int)
    top_wds_p[i] = defaultdict(int)
    
malletpath = '/home/michael/software/mallet-2.0.8RC3/wp15_bl_toks.txt'

with open(malletpath, 'r') as malletfile:
    lines = malletfile.read().splitlines()[1:]
    perspective = ''
    for i, line in enumerate(lines):
        if i % 3 == 0:
            parts = line.split('\t')
            filename = parts[1]
            perspective = filename[-3:-1]
        elif i % 3 == 1:
            parts = line.split(' ')[1:]
            for part in parts:
                item = part.split('/')
                topic = int(item[0])
                wd = item[1]
                if perspective == 'is':
                    top_wds_i[topic][wd] += 1
                elif perspective == 'al':
                    top_wds_p[topic][wd] += 1
        
len(top_wds_i)
```

```python
# Create sorted lists per topic, per perspective
bl_sorted_wds_i = []
bl_sorted_wds_p = []

for topicwds in top_wds_i:
    top_wds = list(reversed(sorted(topicwds.items(), key=operator.itemgetter(1))))
    bl_sorted_wds_i.append(top_wds)

for topicwds in top_wds_p:
    top_wds = list(reversed(sorted(topicwds.items(), key=operator.itemgetter(1))))
    bl_sorted_wds_p.append(top_wds)

for i in range(n_topics):
    print(i)
    print('Israeli:')
    print(', '.join(['{:s}: {:d}'.format(wd, count) for wd, count in bl_sorted_wds_i[i][:25]]))
    print('Palestinian:')
    print(', '.join(['{:s}: {:d}'.format(wd, count) for wd, count in bl_sorted_wds_p[i][:25]]))
    print('')
```

```python
# Print discriminating words 
diff = []
for i in range(n_topics):
    diff.append(defaultdict(int))
    for wd in top_wds_i[i].keys():
        wd_diff = top_wds_i[i][wd] - top_wds_p[i][wd]
        diff[i][wd] = wd_diff

len(diff)
```

```python
# Sort wd differences
sorted_diff = []

for topicwds in diff:
    top_wds = list(sorted(topicwds.items(), key=operator.itemgetter(1)))
    sorted_diff.append(top_wds)
    
for i in range(n_topics):
    print(i)
    print('Palestinian discriminating words:')
    print(', '.join(['{:s}: {:d}'.format(wd, d) for wd, d in sorted_diff[i][:25]]))
    print('Israeli discriminating words:')
    print(', '.join(['{:s}: {:d}'.format(wd, d) for wd, d in bl_sorted_wds_i[i][:25]]))
    print('')
```
