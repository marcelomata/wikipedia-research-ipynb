```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering
from collections import Counter
```

```python
ecpath = '/home/michael/school/research/wp/wikipedia/data/editor_citation_matrix_factors.csv'
ecall = pd.read_csv(ecpath, index_col=0)
ec = ecall[ecall.columns[:-10]]
ec
```

```python
# Initialize hierarchical clustering
hc = AgglomerativeClustering(n_clusters=4)
```

```python
est = KMeans(n_clusters=4, max_iter=1000)
```

```python
# K-means Cluster based on editor-citation values
clusters = est.fit_predict(ec)

# Examine clusters
clusters4 = {i: ec.loc[ec['cluster4'] == i] for i in range(4)}
ec['cluster4'] = clusters
for i in range(4):
    print(len(clusters4[i]))
```

```python
# Hierarchical Cluster based on editor-citation values
clusters = hc.fit_predict(ec)
```

```python
np.savetxt('/home/michael/Desktop/hc_clusters.txt', clusters)
```
