```python
import pandas as pd
from IPython.core.debugger import Tracer; debug_here = Tracer()
```

```python
# Get 100 most frequent citations
freqpath = '/home/michael/school/research/wp/wikipedia/data/citations_annotated.csv'
freq = pd.read_csv(freqpath)
#crit = freq['name'].map(lambda x: isinstance(x, str))
#freq100 = freq[crit]
freq100 = freq.loc[:99,:]
srcs = freq100.sort_values(['name'])['name'].tolist()
srcs
```

```python
itemvecspath = '/home/michael/software/svdfeature-1.2.2/demo/basicMF/itemvecs_10factors.csv'
itemvecs = pd.read_csv(itemvecspath, sep='\s+', header=None, index_col=[0]).sort_index()
itemvecs
```

```python
itemvecs['name'] = srcs
cols = itemvecs.columns[-1:] | itemvecs.columns[:-1]
itemvecs = itemvecs[cols]
itemvecs
```

```python
# Save annotated source vectors
annotated_path = '/home/michael/software/svdfeature-1.2.2/demo/basicMF/itemvecs_10factors_annotated.csv'
itemvecs.to_csv(annotated_path, header=None)
```
