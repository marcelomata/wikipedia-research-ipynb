```python
data = pd.read_csv('/home/michael/school/research/wp/wikipedia/data/talk/ipc_talk_article_urls.csv')
data

# Shuffle threads
thread_inds = defaultdict(list)
for i, row in data.iterrows():
    thread_inds[row['article_title'], row['thread_title']].append(i) 
    
len(thread_inds)

random.sample(thread_inds.items(), 100)
# want to add 2 to get corresponding row in spreadsheet
```
