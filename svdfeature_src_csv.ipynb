```python
src_bias = pd.read_csv('/home/michael/school/research/wp/wikipedia/data/biased_sources.csv')
src_bias
```

```python
src_bias.iloc[1]['name']
```

```python
outrows = []
inpath = '/home/michael/software/svdfeature-1.2.2/demo/basicMF/itemvecs.txt'
with open(inpath) as f:
    for line in f.read().splitlines():
        parts = line.split('\t')
        iid = int(parts[0])
        ivec = [float(val) for val in parts[1].split()]
        row = [src_bias.iloc[iid]['name'], src_bias.iloc[iid]['perspective']] + ivec
        outrows.append(row)
len(outrows)
```

```python
out = pd.DataFrame(outrows)
out.to_csv('/home/michael/software/svdfeature-1.2.2/demo/basicMF/src_vecs.csv', index=False, header=False)
out
```
